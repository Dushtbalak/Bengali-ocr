```
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 
import numpy as np 
import random 
with open("creditcard.csv",'r') as csvfile:
    data1= pd.read_csv(csvfile)
print(data1.columns.values)
print(len(data1))
C = 1
O = 0
data= pd.DataFrame({'Time':data1['Time'], 'Amount':data1['Amount'], 'Class':data1['Class']})
len(data)
data
data.pivot_table(index=['Class'], aggfunc='size')
from sklearn import preprocessing
# creating labelEncoder
le = preprocessing.LabelEncoder()
# Converting string labels into numbers.
label=le.fit_transform(data['Class'])
print(label)
data1=le.fit_transform(data['Time'])
data2=le.fit_transform(data['Amount'])
features=list(zip(data1,data2))

from sklearn.model_selection import train_test_split

# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(features, label, test_size=0.5)

from sklearn.neighbors import KNeighborsClassifier

knn3 = KNeighborsClassifier(n_neighbors=3)

# Train the model using the training sets
knn3.fit(X_train, y_train)
#Predict 
y_pred = knn3.predict(X_test)
predicted= knn3.predict([[100,20]])
print(predicted)
from sklearn import metrics
# Model Accuracy, how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
#line 1-5

#data is the imbalanced data set with minority group(c)
#and majority group(o) of dataframe type
# majority group is the group

nd=[]
knn3data=pd.DataFrame()
knn3data['Time']=data['Time']
knn3data['Amount']=data['Amount']
knn3label=[]
#knn3 is a knn classifier using 3 nearest neighbors for each data in the dataset 
for i in range(0,len(data)):
    re=knn3.predict([[data1[i],data2[i]]])
    knn3label.append(re[0])
    if re[0]==label[i]:
        nd.append("safe")
    else:
        nd.append("noisy")
knn3data['Class']= knn3label
knn3data['Flag']= nd
print(nd)
#C is  1('Not Fraud')
#O is 0('Fraud')
c=0
for i in range(0,len(data)):
    if(nd[i]=='noisy'):
        c=c+1
print(c)
```
